```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv("/kaggle/input/employee-survey/employee_survey.csv")
AgeWorkEnv=df[['Age','WorkEnv']]
AgeJobSatis=df[['Age','JobSatisfaction']]



```

```python
df.isnull()
```

```python
sns.countplot(x='Gender',data=df)
plt.xticks(rotation=45)
plt.show()
```

Much more males than females.

```python
ageenv=df.groupby("Age")['WorkEnv'].mean().reset_index()
plt.figure(figsize = (12,6))
sns.lineplot(x = "Age", y = "WorkEnv",data = ageenv,color="blue",linewidth = 2.5,dashes = False)
plt.title('The Relation Between Age and Work Environment Rating')
plt.xlabel('Age')
plt.xticks(rotation=0)
plt.ylim(1,5)
plt.show()
```

There seems to be the worst work environment rating around the age 55 but increases afterwards.

```python
agebalance=df.groupby("Age")['WLB'].mean().reset_index()
plt.figure(figsize = (12,6))
sns.lineplot(x = "Age", y = "WLB",data = agebalance,color="blue",linewidth = 2.5,dashes = False)
plt.title('The Relation Between Age and Work Life Balance Rating')
plt.xlabel('Age')
plt.ylim(1,5)
plt.show()
```

The Work-Life Balance seems to slightly decrease as the ages progress.

```python
agesatis=df.groupby("Age")['JobSatisfaction'].mean().reset_index()
plt.figure(figsize = (12,6))
sns.lineplot(x = "Age", y = "JobSatisfaction",data = agesatis,color="blue",linewidth = 2.5,dashes = False)
plt.title('The Relation Between Age and Overall Job Satisfaction')
plt.xlabel('Age')
plt.ylim(1,5)
plt.show()
```

Overall job satisfaction dips around 50-57 years but then increases after.

```python
stressleep=df.groupby("SleepHours")['Stress'].mean().reset_index()
plt.figure(figsize = (12,6))
sns.lineplot(x = "SleepHours", y = "Stress",data = stressleep,color="blue",linewidth = 2.5,dashes = False)
plt.title('The Relation Between Sleep Hours and Stress')
plt.xlabel('SleepHours')
plt.ylim(1,5)
plt.show()


```

Stress is highest when sleep hours are less than 4.5.

```python
plt.figure(figsize=(14, 6))
sns.boxplot(x='JobLevel', y='JobSatisfaction', data=df)
plt.title('Job Satisfaction by Job level')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.boxplot(x='Dept', y='JobSatisfaction', data=df)
plt.title('Job Satisfaction by Department')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Dept', hue='Gender', data=df)
plt.title('Gender Distribution by Department')
plt.xlabel('Commute Mode')
plt.show()
```

Majority of male survey participants work in IT, majority of female survey participants work in Finance.

```python
plt.figure(figsize=(14, 6))
sns.boxplot(x='WorkEnv', y='Stress', data=df)
plt.title('Work Life Balance by Department')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='CommuteMode', y='CommuteDistance', data=df)
plt.title('Average Commute Distance by Commute Mode')
plt.xlabel('Commute Mode')
plt.show()
```

People who use a car, public transport, or motorbikes tend to have a greater commute compared to those who bike or walk.

```python
stressbydept=df.sort_values(by='Stress',ascending=False)
stressbydept=stressbydept[['Dept','Stress']]

plt.figure(figsize=(10,6))
sns.barplot(x='Dept', y='Stress', data=df)
plt.title('Average Stress Level by Department')
plt.xlabel('Department')
plt.ylim(1,5)
plt.show()
```

```python
stressbyage=df.sort_values(by='Stress',ascending=False)
stressbyage=stressbyage[['Age','Stress']]

plt.figure(figsize=(10,6))
sns.barplot(x='Age', y='Stress', data=stressbyage)
plt.title('Average Stress Level by Age')
plt.xlabel('Age')
plt.show()
```

Stress levels seems to be evenly spread across the graph, even with the standard error.

**Conclusions**

-The average work life balance slightly decreases as the ages progress.

-Stress tends to rise when employees get an average of 4-4.5 hours of sleep

-IT seems to be a prominent industry for male employees and finance for women out of the industries submitted.

-People who use a car, public transport, or motorbikes tend to have a greater commute compared to those who bike or walk.

-IT employees seem to have the lowest amount of stress out of all the industries submitted.

-The mean for job satisfaction for employees starts to decrease as employees pass the age 52.

-From the tables between people aged 20 to 40 and people aged 41 to 60, the people aged 41 to 60 recieve approximately 18 more hours of training hours per year than people aged 20 to 40. This is a little concerning because younger employees just going into the industry need more training to get ready but they recieve significantly less than older employees according to the data, of which the elders already have many skills from experience.

-The employees aged 41-60 have on average 9 more years of experience than employees aged 20 to 40.

-Older employees can still learn skills, but younger employees require a lot more training compared to older employees who have been in the workforce longer.
